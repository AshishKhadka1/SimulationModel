# Agent

```python
class ChatbotAgent:
    def __init__(self):
        self.knowledge = {
            "hello": "Hello! How can I help?",
            "price": "The price is $100.",
            "refund": "Refund available within 7 days.",
            "bye": "Goodbye!"
        }
        self.reward = 0

    def respond(self, message):
        for key in self.knowledge:
            if key in message.lower():
                return self.knowledge[key], True
        return "I don't understand.", False
```

# Reward

```python
def reward(bot, success):
    if success:
        bot.reward += 2
    else:
        bot.reward -= 1
```

# Main

```python
def main():
    bot = ChatbotAgent()

    for _ in range(5):
        user = input("You: ")
        response, success = bot.respond(user)
        print("Bot:", response)
        reward(bot, success)

    print("Final Reward:", bot.reward)

main()
```
