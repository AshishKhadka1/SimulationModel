# Environment

```python
import random

class Road:
    def __init__(self):
        self.cars = random.randint(0, 10)
```

# Agent

```python
class TrafficAgent:
    def __init__(self):
        self.state = "RED"
        self.reward = 0

    def decide(self, cars):
        if cars > 5:
            self.state = "GREEN"
        else:
            self.state = "RED"

    def act(self, road):
        if self.state == "GREEN":
            road.cars = max(0, road.cars - 2)
            return True
        else:
            road.cars += 1
            return False
```

# Reward

```python
def reward(agent, success):
    if success:
        agent.reward += 3
    else:
        agent.reward -= 1
```

# Main

```python
def main():
    road = Road()
    agent = TrafficAgent()

    for step in range(10):
        agent.decide(road.cars)
        success = agent.act(road)

        print(f"Step {step}: Cars={road.cars}, Light={agent.state}")
        reward(agent, success)

    print("Final Reward:", agent.reward)

main()
```
