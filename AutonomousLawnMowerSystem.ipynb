# Environment

```python
class Lawn:
    def __init__(self, size):
        self.size = size
        self.grid = [['G' for _ in range(size)] for _ in range(size)]  # G = Grass

    def clean(self, x, y):
        if self.grid[x][y] == 'G':
            self.grid[x][y] = '.'
            return True
        return False

    def display(self):
        for row in self.grid:
            print(' '.join(row))
        print()
```

# Agent

```python
class LawnMowerAgent:
    def __init__(self, lawn):
        self.lawn = lawn
        self.position = (0, 0)
        self.reward = 0

    def perceive(self):
        x, y = self.position
        return self.lawn.grid[x][y]

    def act(self):
        x, y = self.position

        if self.perceive() == 'G':
            success = self.lawn.clean(x, y)
            print(f"Cleaning at {self.position}")
        else:
            success = False
            print(f"Moving from {self.position}")

        # Move right
        y = (y + 1) % self.lawn.size
        self.position = (x, y)

        return success
```

# Reward

```python
def reward_agent(agent, success):
    if success:
        agent.reward += 5
    else:
        agent.reward -= 1
```

# Main

```python
def main():
    lawn = Lawn(5)
    agent = LawnMowerAgent(lawn)

    for _ in range(10):
        lawn.display()
        success = agent.act()
        reward_agent(agent, success)

    print("Final Reward:", agent.reward)

main()
```
