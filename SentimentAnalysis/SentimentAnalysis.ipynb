# Sentiment Analysis Sustem

```python
from textblob import TextBlob

# ------------------------------
# 1. SENTIMENT ANALYZER
# ------------------------------
class SentimentAnalyzer:
    def __init__(self):
        """
        Initializes analyzer.
        No training required (TextBlob is pre-trained).
        """
        self.history = []   # store past results
        self.score = 0      # performance score

    def analyze(self, text):
        """
        Analyze sentiment of input text.

        Returns:
        - polarity (-1 to 1)
        - sentiment label
        """

        # Create TextBlob object
        blob = TextBlob(text)

        # Polarity score
        polarity = blob.sentiment.polarity

        # Classification logic
        if polarity > 0:
            sentiment = "Positive"
        elif polarity < 0:
            sentiment = "Negative"
        else:
            sentiment = "Neutral"

        return polarity, sentiment

    def update_score(self, sentiment):
        """
        Reward system based on sentiment type
        (just for learning purpose)
        """
        if sentiment == "Positive":
            self.score += 2
        elif sentiment == "Negative":
            self.score -= 1

    def store_history(self, text, polarity, sentiment):
        """
        Store analysis result
        """
        self.history.append((text, polarity, sentiment))


# ------------------------------
# 2. USER INTERFACE
# ------------------------------
class UserInterface:
    def get_input(self):
        return input("\nEnter text: ")

    def display(self, polarity, sentiment):
        print(f"Polarity Score: {polarity:.2f}")
        print(f"Sentiment: {sentiment}")


# ------------------------------
# 3. MAIN SYSTEM
# ------------------------------
def main():
    print("=== Sentiment Analysis System ===")

    analyzer = SentimentAnalyzer()
    ui = UserInterface()

    for _ in range(5):
        # Step 1: Input
        text = ui.get_input()

        # Step 2: Analyze
        polarity, sentiment = analyzer.analyze(text)

        # Step 3: Display
        ui.display(polarity, sentiment)

        # Step 4: Update score
        analyzer.update_score(sentiment)

        # Step 5: Store history
        analyzer.store_history(text, polarity, sentiment)

    # ------------------------------
    # FINAL REPORT
    # ------------------------------
    print("\n=== SESSION SUMMARY ===")
    print("Final Score:", analyzer.score)

    print("\nHistory:")
    for record in analyzer.history:
        print(record)


# ------------------------------
# RUN PROGRAM
# ------------------------------
if __name__ == "__main__":
    main()
```
